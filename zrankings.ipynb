```python
import requests
from bs4 import BeautifulSoup
import bs4 as bs
import pandas as pd
import json
import urllib.request

```

```python
source = urllib.request.urlopen('https://www.zrankings.com/?page=1').read()
soup = bs.BeautifulSoup(source,'html.parser')
```

```python
best_resorts = []

base_url = "https://www.zrankings.com/?page="

def all_resorts():
    
    page_count = 1
    
    while page_count <= 8:
        url = base_url + str(page_count)
        source = urllib.request.urlopen(url).read()
        
        soup = bs.BeautifulSoup(source,'html.parser')
        
        table_body= soup.find('tbody')
        rows = table_body.findAll('tr',{'class':'top-rowdy'})
        
        for row in rows:
            cols=row.find_all('td')
            cols=[x.text.strip() for x in cols]
            best_resorts.append(cols)
            
        print(page_count)
        page_count +=1 
```

```python
individual_resorts = []
for item in best_resorts:
    resort_name = item[1].split("\n")
    true_snow = item[3]
    acres = item[4]
    vert = item[5]
    summit = item[6]
    lifts = item[7]
    score = item[8]
    
    individual_resorts.append({
        "name": resort_name[0],
        "true_snow": true_snow,
        "acres": acres,
        "vert": vert,
        "summit": summit,
        "lifts": lifts,
        "score": score
    })

df = pd.DataFrame.from_records(individual_resorts)
```

```python
df = df[['name', 'true_snow','acres', 'vert','summit', 'lifts','score']]
```

```python
merged_df['lat'] = merged_df['lat'].astype(float)
merged_df['lng'] = merged_df['lng'].astype(float)
```

```python
def df_to_geojson(merged_df, properties, lat='lat', lon='lng'):
    
    
    geojson = {'type':'FeatureCollection', 'features':[]}

    for _, row in merged_df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}

        feature['geometry']['coordinates'] = [row[lon],row[lat]]

        for prop in properties:
            feature['properties'][prop] = row[prop]
        
        geojson['features'].append(feature)
    
    return geojson
    
    
    
```

```python
useful_columns = ['name','true_snow','acres','vert', 'summit','lifts','score']
geojson_dict = df_to_geojson(merged_df, properties= useful_columns)
geojson_str = json.dumps(geojson_dict, indent=2)
```

```python
output_filename = 'best_resorts.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var dataset = {};'.format(geojson_str))
    
print('{} geotagged features saved to file'.format(len(geojson_dict['features'])))
```
