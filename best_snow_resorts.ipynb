```python
import requests
from bs4 import BeautifulSoup
import bs4 as bs
import pandas as pd
import json
import urllib.request
```

```python
best_snow = []

base_url = "https://www.zrankings.com/ski-resorts/snow?page="

def all_resorts():
    
    page_count = 1
    
    while page_count <= 8:
        url = base_url + str(page_count)
        source = urllib.request.urlopen(url).read()
        
        soup = bs.BeautifulSoup(source,'html.parser')
        
        table_body= soup.find('tbody')
        rows = table_body.findAll('tr')
        
        for row in rows:
            cols=row.find_all('td')
            cols=[x.text.strip() for x in cols]
            best_snow.append(cols)
            
        print(page_count)
        page_count +=1 
```

```python
best_snow_list = []

for item in best_snow:
    resort_name = item[1].split("\n")[0]
    zone = item[2]
    true_snow = item[3]
    data_est = item[4]
    data_elevation = item[5]
    more_six_inches = item[6]
    months_more_90_inches = item[8]
    snowfall_score = item[9]
    north_facing_terrain = item[10]
    south_facing_terrain = item[11]
    total_snow_score = item[12]
    
    best_snow_list.append({
        "name": resort_name,
        "zone": zone,
        "true_snow": true_snow,
        "data_est": data_est,
        "data_elevation": data_elevation,
        "more_six_inches": more_six_inches,
        "90_inches": months_more_90_inches,
        "snowfall_score": snowfall_score,
        "north_facing_terrain": north_facing_terrain,
        "south_facing_terrain": south_facing_terrain,
        "total_snow_score": total_snow_score,
    })
    

df = pd.DataFrame.from_records(best_snow_list)
df = df[['name', 'zone', 'true_snow','data_est','data_elevation','more_six_inches','90_inches', 'snowfall_score','north_facing_terrain','south_facing_terrain', 'total_snow_score']]
```

```python
df2 = pd.read_csv('best_resorts.csv',index_col=0)
df2
```

```python
merged_df = pd.merge(df2, df, on ='name', how = 'right')
merged_df
```

```python
def df_to_geojson(merged_df, properties, lat='lat', lon='lng'):
    
    
    geojson = {'type':'FeatureCollection', 'features':[]}

    for _, row in merged_df.iterrows():
        
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}

        feature['geometry']['coordinates'] = [row[lon],row[lat]]

        for prop in properties:
            feature['properties'][prop] = row[prop]
                geojson['features'].append(feature)
    
    return geojson
    
    
    
```

```python
useful_columns = ['name', 'lat', 'lng', 'true_snow_x', 'acres', 'vert', 'summit', 'lifts',
       'score', 'zone', 'true_snow_y', 'data_est', 'data_elevation',
       'more_six_inches', '90_inches', 'snowfall_score',
       'north_facing_terrain', 'south_facing_terrain', 'total_snow_score']
geojson_dict = df_to_geojson(merged_df, properties= useful_columns)
geojson_str = json.dumps(geojson_dict, indent=2)
```

```python
output_filename = 'best_snow_resorts.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var dataset = {};'.format(geojson_str))
    
print('{} geotagged features saved to file'.format(len(geojson_dict['features'])))
```
